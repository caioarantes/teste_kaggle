```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("emmarex/plantdisease")

print("Path to dataset files:", path)
```

```python
import os

# List all files in the downloaded dataset directory
all_files = []
for root, _, files in os.walk(path):
    for f in files:
        all_files.append(os.path.join(root, f))

print(f"Total files found: {len(all_files)}")
for file_path in all_files[:20]:  # preview first 20
    print(file_path)

# Open a few files to verify access
for file_path in all_files[:5]:
    with open(file_path, "rb") as file:
        header = file.read(32)
    print(f"\nOpened: {file_path}\nFirst 32 bytes: {header}")
```
